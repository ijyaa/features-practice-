## Categorizing Variables:

- Country: Categorial 
- Year: Catergorical 
- Total TV Watched on Average: Continious Variable (Ratio Variable) 

Ratio Variable because it can have a rank, distance and a meaningful zero value. Example, someone can watch 0 mins of TV (none) and people that watch TV for 10 hrs watch double of the TV from someone that watches 5 hrs. 
- People's Trust(Most ppl can be trusted or not): Continious Variable (Ordinal)

Ordinal because it indicates a ranking system (High Trust, Low Trust, No Trust) etc but, it does not give any other meaningful information such as rank order or a good way to measure difference between trust levels from one person to another 

- Fairness (Most people are fair or not): Continious Variable (Ordinal)

Ordinal because it indicates a ranking system (High Fairness, Low Fairness, No Fairness) etc but, it does not give any other meaningful information such as rank order or a good way to measure difference between fairness levels from one person to another. 

- Helpful (Most people are helpful or not): Continious Variable (Ordinal)

Ordinal because it indicates a ranking system (High Helpfulness, Low Helpfulness, No Help) etc but, it does not give any other meaningful information such as rank order or a good way to measure difference between helpfulness levels from one person to another. 


- Happiness index: Continious Variable (Ordinal Variable)

I would say interval variable because happiness is variable subjective in nature and we cannot measure if level 10 happiness for someone means the same to the other person. But, there is a rank system in ways people they say example: Very Happy, Less Happy etc. 

- Social Meeting Frequency: Continious Variable (Ratio Variable) 

This has a rank, distance, and also a zero point. Someone might not see anyone versus someone who sees their friends 4 times a month sees their friends double compared to someone who only sees them twice a month. There is a zero point as well since people who see people 0 times a month don't see them at all. 

- Gender: Categorical 

- Age: Categorical 

- Partner: Categorical

```python
# Reverse-coding the 'happy' variable so that higher scores mean less
# happiness (and thus possibly more sadness).

features['Sadness'] = max(df['happy']) - df['happy']
```

```python
sns.regplot(
    df['tvtot'],
    features['Sadness'], 
    x_jitter = .49, 
    y_jitter = .49, 
    scatter_kws= {'alpha': 0.3}, 
    line_kws = {'color': 'black'}, 
    ci = None)

plt.xlabel('TV Watching')
plt.ylabel('Sadness')
plt.show()
```
